# Tabular EDA Template
Check target balance, missing values and leakage.

```
import pandas as pd
df = pd.read_csv('train.csv')
missing = df.isna().sum().sort_values(ascending=False)
target_balance = df['target'].value_counts(normalize=True)
suspect = [c for c in df.columns if 'target' in c.lower() or 'label' in c.lower()]
```

```
numeric_cols = df.select_dtypes(include=['number', 'bool']).columns
corr = df[numeric_cols].corr(numeric_only=True)['target'].sort_values(key=lambda s: s.abs(), ascending=False)
```
